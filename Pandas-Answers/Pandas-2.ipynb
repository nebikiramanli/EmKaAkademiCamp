```python
import pandas as pd
import numpy as np

df_tips = pd.read_csv('./tips.csv')
df_tips.head()
```

```python
# Q-1 
# 1) df_tips içindeki Payment ID column'unun son 3 rakamını çekin. Bu işlemi 3 farklı yolla da yapmaya çalışın.

print("1. Yol: \n", df_tips['Payment ID'].str[-3:])
print("2. Yol: \n", df_tips['Payment ID'].str.slice(-3))
print("3. Yol: \n", df_tips['Payment ID'].apply(lambda x: x[-3:]))
```

```python
# Q-2
# 2) df_tips içindeki Payer Name sütununda isim ve soy isim bilgileri bulunuyor. 
# O sütundan sadece isim bilgileri çekin ve 'JustName' adında yeni bir column oluşturun. 
# (Bu biraz zorlayıcı olabilir, Önce kendiniz yapmaya çalışın fakat yapamazsanız 
# ChatGPT'ye soruyu düzgünce sorarak cevabını kod olarak alabilirsiniz, ardından döndürülen kodu anlamaya çalışmanız yeterlidir.)

df_tips['JustName'] = df_tips['Payer Name'].apply(lambda x: x.split()[0])
df_tips.head()
```

```python
# Q-3
# 3) 2.Soruda yeni oluşturduğunuz 'JustName' sütunundaki verilerin her birinden kaç tane olduğuna uygun methodu kullanarak bakın.

df_tips['JustName'].value_counts()
```

```python
# Q-4
# 4) 3.Soruda gördüğünüz gibi JustName sütununda 'Jason' isminden 9 tane, 'Michael' isminden 7 tane , 'James' isminden 6 tane bulunuyor. 
# Peki bunların hepsi gerçekten farklı kişiler mi ? Hemen 'Payer Name' sütunundaki verilerin adetlerini de saydırtarak bir kıyaslama yapın. ,
# 'JustName' sütununu kullanmanın faydasız ve yanıltıcı olabileceğini görün ardından oluşturduğumuz 'JustName' sütununu kalıcı olarak drop edin.
# (Oluşturmanız için bir pratikti bu sütun sadece :D)

print("Eşssiz", df_tips['Payer Name'].value_counts())
print("JustName", df_tips['JustName'].value_counts())


```

```python
# Soru 4 Devamı
df_tips.drop('JustName', axis=1, inplace=True)
```

```python
# Q-5
# 5) Erkeklerin ve kadınların davranış patternlerini ayrı ayrı inceleyip yorumda bulunabilmek için 'df_male' ve 'df_female' adında 2 dataframe oluşturun.
#  Birinde sadece 157 erkek verisi, diğerinde sadece 87 kadın verisi bulunsun.

df_male = df_tips[df_tips["sex"] == "Male"][0:157]
df_female = df_tips[df_tips["sex"] == "Female"][0:87]

print("Male \n", df_male)
print("Female \n", df_female)
```

```python
# Q-6
# 6) Ardından bu 'df_male' ile 'df_female' 'in ödedikleri bahşişler toplamını bulup karşılaştırın.

male_tips_sum = df_male['tip'].sum()
female_tips_sum = df_female['tip'].sum()

print("Male Tips Sum   :", male_tips_sum)
print("Female Tips Sum :", female_tips_sum)
```

```python
# Q-7
# 7) 6.Soruda erkeklerin ve kadınların ödedikleri bahşişler toplamına bakıldığında 
# Erkeklerin kadınlara kıyasla toplamda daha çok bahşiş ödediğini görüyoruz. 
# Fakat biz biliyoruz ki bizim verimizde 157 Erkek, 87 kadın verisi vardı. 
# Dolayısıyla Erkek sayısı daha fazla olduğu için toplam ödeyecekleri bahşiş miktarlarında da daha fazla çıkması oldukça normal bir durum. 
# Bu yüzden daha mantıklı bir çıkarım için bu sefer df_male ile df_female'in bahşiş ORTALAMALARIN'a bakıp kıyaslayın, aralarında anlamlı ve ciddi bir fark var mı ?

male_tips_mean = df_male['tip'].mean()
female_tips_mean = df_female['tip'].mean()

print("Male Tips Mean   :", male_tips_mean)
print("Female Tips Mean :", female_tips_mean)
```

```python
# Q-8
# 8) Oluşturmuş olduğunuz df_male ile çalışarak elimizdeki verideki erkeklerin yüzde kaçının sigara içtiğini bulun. 
# Aynı şekilde elimizdeki kadınların yüzde kaçının sigara içtiğini de bulun. Aralarında anlamlı bir fark mı yorumlayın.
smoker_male = df_male[df_male["smoker"] == "Yes"]
smoker_male_count = smoker_male['smoker'].count()

print("Male smoker rate : % ", (smoker_male_count / df_male['smoker'].count()) * 100)
```

```python
# Q-9
# 9) Erkeklerin zaman olarak dinnerda mı yoksa lunch'da mı daha fazla yemek yediğini yüzdesel olarak bulun. 
# Aynı işlemi kadınlar için de yapın. 
# Yüzdesel olarak erkeklerin ve kadınların yemek yeme vakitleri arasında anlamlı bir fark var mı yok mu yorumlayın.

male = df_tips[df_tips["sex"] == "Male"]
male_time_group = male.groupby('time').count()
male_dinner_rate = male_time_group.loc['Dinner'] / male_time_group.sum() * 100
print("Male Dinner Rate : \n% ", male_dinner_rate.iloc[0].round(2))
male_lunch_rate = male_time_group.loc['Lunch'] / male_time_group.sum() * 100
print("Male Lunch Rate : \n% ", male_lunch_rate.iloc[0].round(2))

female = df_tips[df_tips["sex"] == "Female"]
female_time_group = female.groupby('time').count()
female_dinner_rate = female_time_group.loc["Dinner"] / female_time_group.sum() * 100
print("Female Dinner Rate : \n%", female_dinner_rate.iloc[0].round(2))
female_lunch_rate = female_time_group.loc["Lunch"] / female_time_group.sum() * 100
print("Female Lunch Rate : \n%", female_lunch_rate.iloc[0].round(2))
```

```python
# Q-10
# 10) CC Number sütununu kalıcı olarak drop edin. 
# Ardından df'in içindeki sütunların birbirleriyle olan ilişkilerini inceleyin ve çıkan sonuçlar hakkında yorumlarınızı belirtin.

df_tips.drop('CC Number', axis=1, inplace=True)

print(df_tips.corr())
```
