```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./mpg.csv")
```

```python
# Q-1 
# df'in içindeki sütunların birbirleriyle olan korelasyonlarına(ilişkilerine) bakın.

corr = df.corr()
corr 
```

```python
# Q-2
# mpg sütunu ile diğer sütunların arasındaki korelasyonları özel olarak inceleyin.
# (Bunu göstermedik ama df.corr() çağırdıktan sonra aynı normal df'lerde olduğu gibi istediğiniz sütunu çağırabiliyorsunuz.)

corr_mpg = df.corr()['mpg']
corr_mpg
```

```python
# Q-3
#  cylinders sütununda hangi unique değerler olduğuna liste halinde bakın.

df['cylinders'].unique()
```

```python
# Q-4
# 4) cylinders sütunundaki unique değerlerin her birinden kaçar adet olduğuna bakın.

df['cylinders'].value_counts()
```

```python
# Q-5
# 5) Arabanın modeli 75-80 yılları arasındaki arabaların hepsini getirin.(75 ve 80 dahil)

df[(df['model_year'] >= 75) & (df['model_year'] <= 80)]
```

```python
# Q-6
# 6) horsepower'ı 155-180 arasındaki arabaların hepsini getirin.
# (155 dahil, 180 dahil değil)(Bu sorguyu yapabilmeniz için horsepower sütununu object durumdan float duruma getirmeniz gerekiyor.)

df['horsepower'] = df['horsepower'].replace('?', np.nan)
df['horsepower'] = df['horsepower'].astype('float')
df[(df['horsepower'] >= 155) & (df['horsepower'] < 180)]
```

```python
# Q-7
# 7) Elinizdeki df'i horsepower'a göre büyükten küçüğe sıralayın ardından df'in indexlerini sıfırlayın.

df.sort_values(by='horsepower', ascending=False).reset_index(drop=True)
```

```python
# Q-8
# 8) Elinizdeki df'den en küçük 5 acceralation'a sahip olan arabaları getirin.

df.nsmallest(5, 'acceleration')
```

```python
# CIA Country Facts
# --- CIA_Country_Facts.csv dosyasını okuyun ve yeniden df değişkenine atayın.(Artık eski df ile işimiz kalmadığı için aynı ismi verebiliriz.)

df = pd.read_csv("./CIA_Country_Facts.csv")
```

```python
# Q-9
# 9) Elimizdeki CIA'in ülkeler hakkında topladığı basit açık kaynak verileri df'inde kaç ülkenin verisi olduğunu bulun.

df['Country'].count()
```

```python
# Q-10
# 10) Elimizdeki sütunlardan hangilerinde kaçar tane nan veri olduğunu bulun.

df.isna().sum()
```

```python
# Q-11
# 11) Elimizde tamamen nan değerlerden oluşan bir satır varsa o satırı düşürün.
# (Tamamen nan değerlerden oluşan bir satır olup olmadığını, herhangi bir satırın düşüp düşmediğini kontrol edin)

df.dropna(how='all', inplace=True)
df.isna().sum()
```

```python
# Q-12
# 12) Her Bölgedeki(Region) toplam kişi sayısına bakın.

df.groupby('Region')['Population'].sum()
```

```python
# Q-13
# 13) Climate değerleri nan olanlar, tüm df'in ortalama climate değeriyle doldurun.
# (Gerçek bir analiz yapıyor olsaydık bu çok mantıklı bir hareket olmazdı, 
# belki her bulunduğu bölgenin ortalama climate değeriyle doldurmak daha mantıklı bir hareket olurdu. Fakat şimdilik bu kadar kasmaya gerek yok. 
# nan değerleri tüm climate ortalamalarıyla doldursanız yeterli.)

df['Climate'].fillna(df['Climate'].mean(), inplace=True)
```

```python
# Q-14
# 14) Bölgeler ve Ülkelere göre gruplama yapın ardından Doğum miktarlarına bakın.

df.groupby(['Region', 'Country'])['Birthrate'].sum()    
```

```python
# Q-15
# 15) Veriyi istediğiniz şekilde inceleyerek bu zamana kadar öğrendiğiniz bilgilerle anlamlı 5 adet yargı cümlesi çıkartın. 
# ( En çok insan Asya kıtasında yaşıyor, En çok insan Afrinın sahralarında ölüyor, Ortalama en yüksek telefon sahibi Kuzey Amerikada, 
# En çok okuma yazma oranına sahip ülkeler Baltık ülkeleri vb.)

print(df.groupby('Region')['Population'].sum().sort_values(ascending=False).head(1))
```
